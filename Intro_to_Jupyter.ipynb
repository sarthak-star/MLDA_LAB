## Python code to calculate Factorial of a given number 

```python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n-1)
```

```python
# Test the factorial function
print(factorial(5))  # Example with input 5
print(factorial(0))  # Example with input 0
```

## Import a Python library (e.g., NumPy) and use it to perform a mathematical operation.

```python
import numpy as np

# Example: Calculate the mean and median of a list of numbers using NumPy
numbers = [2, 5, 8, 12, 15]
mean_value = np.mean(numbers)
median_value = np.median(numbers)

print("Mean:", mean_value)
print("Median:", median_value)
```

## Visualize data using Matplotlib or Seaborn libraries within a Jupyter Notebook.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Example: Create a simple bar chart using Matplotlib
categories = ['A', 'B', 'C', 'D']
values = [25, 40, 30, 45]

plt.bar(categories, values)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart Example')
plt.show()

# Example: Create a scatter plot using Seaborn
data = sns.load_dataset('iris')
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=data)
plt.title('Scatter Plot Example')
plt.show()
```

## Understanding Kernels

Different file formats of Jupyter Notebook serve different purposes and may be useful in various scenarios:

.ipynb (IPython Notebook):

This is the default file format for Jupyter Notebook.
Useful for creating interactive documents containing live code, equations, visualizations, and explanatory text.
Ideal for sharing notebooks with colleagues, collaborators, or posting them on platforms like GitHub.
.html (HTML):

Exporting a notebook to HTML format can be useful for creating static, self-contained versions of the notebook.
Useful for sharing notebooks with individuals who may not have Jupyter Notebook installed, as HTML files can be viewed in any web browser.
.pdf (PDF):

Exporting a notebook to PDF format is helpful when creating printable, formatted documents.
Useful for sharing finalized reports, presentations, or academic papers with a wider audience, as PDF files maintain consistent formatting across different platforms.
.py (Python script):

Converting a notebook to a Python script (.py) is beneficial when you want to distribute the code without the narrative or visualizations.
Useful for software development workflows, version control systems, or automation tasks where a pure Python script is required.
.md (Markdown):

Exporting a notebook to Markdown format can be useful for integrating code and narrative into documentation, blogs, or websites.
Markdown files are lightweight, easy to read, and can be easily converted to various formats using markdown processors.
Each file format has its own advantages and use cases, and the choice depends on the intended audience, purpose of sharing, and the level of interactivity or formatting required in the document.

##  interactive widgets in Jupyter 

```python
import ipywidgets as widgets
from IPython.display import display
from ipywidgets import interact, IntSlider
```

```python


slider = widgets.IntSlider(
    value=5,  # Initial value
    min=0,    # Minimum value
    max=10,   # Maximum value
    step=1,   # Step size
    description='Number:', 
    orientation='horizontal'  
)

def plot_sine_wave(frequency):
    x = np.linspace(0, 10, 1000)  
    y = np.sin(2 * np.pi * frequency * x)  

    # Plot the sine wave
    plt.figure(figsize=(8, 6))
    plt.plot(x, y, color='blue', linewidth=2, linestyle='-')
    plt.title(f'Sine Wave (Frequency: {frequency} Hz)')
    plt.xlabel('Time')
    plt.ylabel('Amplitude')
    plt.grid(True)
    plt.show()

interact(plot_sine_wave, frequency=slider)
```
