### Q2) A program for linear regression model using scikit-learn but no machine learning

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import pandas as pd
from sklearn.model_selection import train_test_split

# Generate some random data for demonstration
df = pd.read_csv("advertising.csv")
print(df.head())



X = df[['Radio']]
y = df["Sales"]

X_train , X_test , y_train , y_test = train_test_split(X,y,test_size = 0.2)
```

```python
# Create a Linear Regression model
model = LinearRegression()

# Fit the model (train)
model.fit(X_train, y_train)

val_predictions = model.predict(X_test)

# Plotting the data and the linear regression line
plt.plot(X_test, val_predictions, color='red', linewidth=2)
plt.xlabel('X')
plt.ylabel('y')
plt.title('Linear Regression using scikit-learn (without ML)')
plt.show()
```
